### Calculate the average ERA for each team, list the teams by the average ERA in a descending order.


```python
df_era = pitching_df[['playerID', 'nameFirst', 'nameLast', 'ERA', 'yearID', 'teamID']]
df_era_groupby_team = df_era.groupby('teamID')[['ERA']].mean()
df_era_groupby_team.sort_values('ERA', ascending = False)
```

### Calculate the total number of home runs allowed (HRs) for each team, and list the team by the count in an ascending order

```python
df_team = pitching_df[['teamID' , 'HR']].groupby('teamID').sum().sort_values('HR', ascending = True)
df_team
```

### Calculate each team’s average ERA from 2011 to 2015, and list the result as a table, where each column is a year, each row is a team

```python
year_df = df_era[(df_era['yearID']>=2011) & (df_era['yearID']<=2015)]
avg_era_df = year_df.groupby(['teamID', 'yearID'], as_index = True)[['ERA']].mean().T
avg_era_df
```

### Calculate the average ERAs for American League (AL) and National League (NL) from 2011 to 2015, and list the result as a table, where each column is a league and each row is a year.

```python
league_df = pitching_df[['ERA', 'yearID', 'teamID','lgID']]
specific_league_df = league_df[((league_df['yearID']>=2011) &
                                (league_df['yearID']<=2015) &
                                (league_df['lgID']=='AL') |
                                (league_df['lgID']=='NL'))]
specific_league_df = specific_league_df.groupby(['lgID', 'yearID'], as_index = True)[['ERA']].mean().T
specific_league_df
```

### List the pitchers who had the most wins in each of the years from 2011 to 2015.

```python
total_pitcher_df = pitching_df[['playerID', 'nameFirst', 'nameLast', 'W', 'yearID']]
specific_year_pitcher_df = total_pitcher_df[(total_pitcher_df['yearID']>=2011) & (total_pitcher_df['yearID']<=2015)]
specific_year_pitcher_df.sort_values('W', ascending = False)
```
